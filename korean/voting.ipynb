```python
import csv
```

```python
file_dir = './data/'
file_list = [
    'ko_sample.csv', 
    'bert_batch32_epoch5.csv', 
    'bert_batch32_epoch10.csv',
    'bert_with_test_batch32_epoch4.csv',
    'bert_with_test_batch32_epoch5.csv',
    'hanbert_with_test_epoch7.csv'
]
input_data = 'ko_data.csv'
result_file = 'last_vote_result.csv'
```

```python
# test dataset으로부터 id 목록을 list로 받아온다
def find_ids(file_name):
    ids = []
    with open(file_name, 'r', encoding='CP949') as f:
        rdr = csv.reader(f)
        for line in rdr:
            ids.append(line[0])
    return ids
```

```python
# id list를 통해 result 값을 저장할 dictionary를 만든다
def init_dict(id_list):
    result = dict()
    for id_val in id_list:
        result[id_val] = []
    return result
```

```python
# csv 결과 파일 하나를 받아와서 list로 저장한다
def get_result(file_name):
    contents = []
    with open(file_name, 'r', encoding='CP949') as f:
        rdr = csv.reader(f)
        for line in rdr:
            contents.append(line)
    return contents
```

```python
# 각 id에 대해 최빈값을 선택한다
def count_korean(lst):
    zero = 0
    one = 0
    for l in lst:
        if l == '0':
            zero += 1
        elif l == '1':
            one += 1
        else:
            print("Error! {} should not be included".format(l))
    # 만약 zero / one의 개수가 같은 경우는 어떻게 할까에 대한 고민 필요!
    if zero > one:
        return 0
    else:
        return 1
```

```python
# id 목록과 dictionary 결과를 바탕으로 최종 결과를 선택한다
def vote_korean(id_list, dictionary):
    result = []
    for id_val in id_list[1:]:
        expectations = dictionary[id_val]
        result.append([id_val, count_korean(expectations)])
    return result
```

```python
def make_result_file(file_name, result):
    with open (file_name, 'w') as f:
        wr = csv.writer(f)
        wr.writerow(['Id', 'Predicted'])
        for r in result:
            wr.writerow(r)
```

```python
def find_result(idx, lst):
    for l in lst:
        if l[0] == idx:
            return l[1]
    return 'Not Found'
```

```python
# dataset의 id 값들을 모두 받아온다
id_list = find_ids(file_dir + input_data)
# id 값들을 바탕으로 dictionary를 초기화한다
result = init_dict(id_list)

# 각 결과 파일들의 예측값을 수합한다
for file_name in file_list:
    res = get_result(file_dir + file_name)
    for r in res:
        [input_id, expectation] = r
        result[input_id].append(expectation)
# 수합된 예측값을 바탕으로 최종 결과를 도출한다
vote_result = vote_korean(id_list, result)

# Test Code #
sample_id = id_list[1:11]
for sample in sample_id:
    vote = result[sample]
    res = find_result(sample, vote_result)
    print(file_name)
    print("For id '{}', {} was selected from {}".format(sample, res, vote))

make_result_file(file_dir + result_file, vote_result)
```
